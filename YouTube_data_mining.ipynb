```python
# First round data extraction
# Get information from title and descriptions
# Output: ./YouTube_dataset/YouTube_dataset_version1.csv 
import pandas as pd
from py_trans import PyTranslator
import emoji as emo
from nltk.tokenize import RegexpTokenizer
from string import punctuation
import sys
import subprocess
from tqdm import tqdm
from langdetect import detect

df = pd.read_csv('./YouTube_dataset/YouTube_dataset.csv')
print(df)
titles = df['title'].to_list()
descriptions = df['description'].to_list()

tokenizer = RegexpTokenizer(r'\w+')
t_emoji_count, t_word_count, t_punctuation_count, t_length, t_lang= [], [], [], [], []
d_emoji_count, d_word_count, d_punctuation_count, d_length, d_lang= [], [], [], [], []
for i, _ in enumerate( tqdm(titles) ):
    # Title search
    titles[i] = str(titles[i])
    if len(titles[i]) > 0:
        t_emoji_count.append(emo.emoji_count(titles[i])) #------------------------------------------
    else:
        t_emoji_count.append(0)
    if t_emoji_count[i] == 0:
        temp_str = titles[i]
    else:
        temp_str = emo.replace_emoji(titles[i], replace = "") # remove all emojis
    try: t_lang.append(detect(temp_str)) #------------------------------------------
    except: t_lang.append('') #------------------------------------------    
    temp_list = tokenizer.tokenize(temp_str) 
    t_word_count.append(len(temp_list)) #------------------------------------------
    temp_punc = 0
    for c in titles[i]:
        if c in punctuation: temp_punc += 1
    t_punctuation_count.append(temp_punc) #------------------------------------------
    t_length.append(len(titles[i])) #------------------------------------------
    # Description search 
    descriptions[i] = str(descriptions[i])
    if len(descriptions[i]) > 0:
        d_emoji_count.append(emo.emoji_count(descriptions[i])) #------------------------------------------
    else:
        d_emoji_count.append(0)  
    if d_emoji_count[i] == 0:
        temp_str = descriptions[i]
    else:
        temp_str = emo.replace_emoji(descriptions[i], replace = "") # remove all emojis
    try: d_lang.append(detect(temp_str)) #------------------------------------------
    except: d_lang.append('') #------------------------------------------
    temp_list = tokenizer.tokenize(temp_str)
    d_word_count.append(len(temp_list)) #------------------------------------------
    temp_punc = 0
    for c in descriptions[i]:
        if c in punctuation: temp_punc += 1
    d_punctuation_count.append(temp_punc) #------------------------------------------
    d_length.append(len(descriptions[i])) #------------------------------------------
print('done')
# print(t_emoji_count)
# print(t_word_count)
# print(t_punctuation_count)
# print(t_length)
# print(d_emoji_count)
# print(d_word_count)
# print(d_punctuation_count)
# print(d_length)
df['t_emoji_count'] = t_emoji_count
df['t_word_count'] = t_word_count
df['t_punctuation_count'] = t_punctuation_count
df['t_length'] = t_length
df['t_lang'] = t_lang
df['d_emoji_count'] = d_emoji_count
df['d_word_count'] = d_word_count
df['d_punctuation_count'] = d_punctuation_count
df['d_length'] = d_length
df['d_lang'] = d_lang
df.drop(columns = ['Unnamed: 0','Unnamed: 0.1'])
df.to_csv('./YouTube_dataset/YouTube_dataset_version1.csv')     
print(df)
```

```python
# Second round data extracting
# Get information from Thumbnails
# Output: ./YouTube_dataset/YouTube_dataset_version2.csv 
import pandas as pd
from PIL import Image
import requests
import matplotlib.pyplot as plt
import numpy as np
from tqdm import tqdm
from math import sqrt
import cv2
from scipy.stats import entropy

def getBlurValue(image):
    canny = cv2.Canny(image, 50,250)
    return np.mean(canny)

df = pd.read_csv('./YouTube_dataset/YouTube_dataset_version1.csv')
thumbnails = df['thumbnail'].to_list() 
# r = requests.get(url, stream=True)
# img = Image.open(r.raw)
# # Convert thumbnail to np array
# img_np = np.asarray(img)
# print(img_np)
# # Convert np array to thumbnail
# im = Image.fromarray(np.uint8(img_np))
# plt.imshow(img)
# plt.show()
# plt.imshow(im)
# plt.show()
thumb_brightness = []
thumb_contrast = []
thumb_blurriness = []
thumb_entropy = []


for i, _ in enumerate( tqdm(thumbnails) ):
    ## The following code is used to get thumbnail from the corresponding url 
    # Thumbnails search
    r = requests.get(thumbnails[i], stream=True)
    img = Image.open(r.raw)
    img_np = np.asarray(img)
    (Nx, Ny, _) = np.shape(img_np)
    p_brightness = []
    # Compute Brightness
    for r in img_np:
        for p in r:
            p_brightness.append(0.299 * p[0] + 0.587 * p[1] + 0.114 * p[2])        
    brightness = (1/(765 * Nx * Ny) * sum(p_brightness))
    # Compute Contrast
    p_contrast = []
    for p in p_brightness:
        p_contrast.append((p - brightness)**2)
    contrast = sqrt((1/(765 * Nx * Ny) * sum(p_contrast)))
    # Compute Entropy
    p = np.histogram(img_np, bins=256, range=(0,256), density=True)[0]
    en = entropy(p, base=2)
    
    # Add data to the table
    thumb_brightness.append(brightness)
    thumb_contrast.append(contrast)
    thumb_blurriness.append(getBlurValue(img_np))
    thumb_entropy.append(en)
#     if i%100 == 0:
#         plt.imshow(img)
#         plt.show()
#         print(img)

print('done')
df['thumb_brightness'] = thumb_brightness
df['thumb_contrast'] = thumb_contrast
df['thumb_blurriness'] = thumb_blurriness
df['thumb_entropy'] = thumb_entropy
df.to_csv('./YouTube_dataset/YouTube_dataset_version2.csv')     
print(df)
```

```python
# Normalize the extracted data - use zscore
import pandas as pd
from scipy.stats import zscore

df = pd.read_csv('./YouTube_dataset/YouTube_dataset_version2.csv')



# Map languages to digit space
t_lang = df['t_lang']
d_lang = df['d_lang']
t_lang_set = list(set(t_lang))
d_lang_set = list(set(d_lang))
t_lang_value = list(range(0,len(t_lang_set)))
d_lang_value = list(range(0,len(d_lang_set)))
t_lang_digit, d_lang_digit = [], []
for i in range(len(t_lang)):
    val = t_lang_value[t_lang_set.index(t_lang[i])]
    t_lang_digit.append(val)
for i in range(len(d_lang)):
    val = d_lang_value[d_lang_set.index(d_lang[i])]
    d_lang_digit.append(val)

df['t_lang_digit'] = t_lang_digit
df['d_lang_digit'] = d_lang_digit


all_num_data = [
    df['t_emoji_count'], df['t_word_count'], df['t_punctuation_count'], df['t_length'], df['t_lang_digit'],
    df['d_emoji_count'], df['d_word_count'], df['d_punctuation_count'], df['d_length'], df['d_lang_digit'],
    df['thumb_brightness'], df['thumb_contrast'], df['thumb_blurriness'], df['thumb_entropy']
]
# feature1 - feature14
for i in range(len(all_num_data)):
    df['feature'+str(i+1)] = zscore(all_num_data[i])
df.to_csv('./YouTube_dataset/YouTube_dataset_version3.csv')     
print(df)
    
```
